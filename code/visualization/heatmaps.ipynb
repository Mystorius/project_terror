```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

import pandas as pd
import numpy as np
import os
```

```python
np.set_printoptions(threshold=np.nan)
#pd.set_option('display.height', 1000)
pd.set_option('display.max_rows', 2000)
pd.set_option('display.max_columns', 500)
#pd.set_option('display.width', 1000)
```

```python
#creates heatmap for a predefined year, heat is measured on number of kills per city
#load data and basic manipulation
df = pd.read_csv(r"data_clean.csv", sep="	")
df["year"] = df['start_date'].astype(str).str[:4]
df = df.dropna(subset=['latitude', 'longitude'])

#filter for year
df_year = df.loc[df['year'] == '2010'] #define year
df_year = df_year[["city","latitude","longitude","nkill"]]
df_year = df_year.dropna(subset=['latitude', 'longitude',"city"])
df_year = df_year[df_year.city != 'Unknown']

df_add = pd.DataFrame() 
df_add["city"]=""
df_add["latitude"]=""
df_add["longitude"]=""

#each attack in one city has the same coordinates so we have to add up total number of kills, to determine heat
#df_add is dataframe with the columns city, latitude, longitude, score
cities = df_year['city'].unique() 

for c in cities:
    a = pd.DataFrame({"city": [df_year.loc[df_year['city'] == c]["city"].iloc[0]],
             "latitude": [df_year.loc[df_year['city'] == c]["latitude"].iloc[0]],
             "longitude": [df_year.loc[df_year['city'] == c]["longitude"].iloc[0]]})
    df_add = df_add.append(a)
    
df_add["nkill"]=""
df_add = df_add.set_index('city')    
cities = df_add.index.values
for c in cities:
    df_add.at[c, 'nkill'] = df_year.loc[df_year['city'] == c]["nkill"].sum()
df_add = df_add.reset_index()


df_add["score"]=""# determines score for each city for size and colour in the map
df_add.loc[df_add["nkill"] == 0,"score"] = 1000
df_add.loc[(df_add['nkill'] > 0) & (df_add['nkill'] <= 3),"score"] = 1500
df_add.loc[(df_add['nkill'] > 3) & (df_add['nkill'] <= 5),"score"] = 1800
df_add.loc[(df_add['nkill'] > 5) & (df_add['nkill'] <= 10),"score"] = 3000
df_add.loc[(df_add['nkill'] > 10) & (df_add['nkill'] <= 20),"score"] = 4500
df_add.loc[(df_add['nkill'] > 20) & (df_add['nkill'] <= 50),"score"] = 6000
df_add.loc[(df_add['nkill'] > 50) & (df_add['nkill'] <= 80),"score"] = 7000
df_add.loc[(df_add['nkill'] > 80) & (df_add['nkill'] <= 100),"score"] = 8000
df_add.loc[(df_add['nkill'] > 100) & (df_add['nkill'] <= 150),"score"] = 9000
df_add.loc[(df_add['nkill'] > 150) & (df_add['nkill'] <= 250),"score"] = 10000
df_add.loc[(df_add['nkill'] > 250) & (df_add['nkill'] <= int(df_add.nkill.max())),"score"] = 12000

#colour could be used with a different score an then determine a colourcode for each class
#df_add[""]=""
#df_add.loc[df_add["nkill"] == 0,"colour"] = 1
#df_add.loc[(df_add['nkill'] > 0) & (df_add['nkill'] <= 3),"colour"] = 2
#df_add.loc[(df_add['nkill'] > 3) & (df_add['nkill'] <= 5),"colour"] = 3
#df_add.loc[(df_add['nkill'] > 5) & (df_add['nkill'] <= 10),"colour"] = 4
#df_add.loc[(df_add['nkill'] > 10) & (df_add['nkill'] <= 20),"colour"] = 5
#df_add.loc[(df_add['nkill'] > 20) & (df_add['nkill'] <= 50),"colour"] = 6
#df_add.loc[(df_add['nkill'] > 50) & (df_add['nkill'] <= 80),"colour"] = 7
#df_add.loc[(df_add['nkill'] > 80) & (df_add['nkill'] <= 100),"colour"] = 8
#df_add.loc[(df_add['nkill'] > 100) & (df_add['nkill'] <= 150),"colour"] = 9
#df_add.loc[(df_add['nkill'] > 150) & (df_add['nkill'] <= 250),"colour"] = 10
#df_add.loc[(df_add['nkill'] > 250) & (df_add['nkill'] <= int(df_add.nkill.max())),"score"] = 11

#necessary parameters for map
lat = df_add['latitude'].values
lon = df_add['longitude'].values
size = df_add['score'].values
#col = df_add['colour'].values

#map
fig = plt.figure(figsize=(75, 75), edgecolor='w')
m = Basemap(projection='cyl', resolution=None,
            llcrnrlat=-90, urcrnrlat=90,
            llcrnrlon=-180, urcrnrlon=180, )
m.shadedrelief()

#scatter on map
m.scatter(lon, lat, 
          latlon = True, 
          c = size, #color, sequence, or sequence of color
          s = size,# #scalar or array_like, shape (n, ), optional
          cmap='Reds',
          alpha=0.5) # The alpha blending value, between 0 (transparent) and 1 (opaque).

#title
plt.title("1971", fontsize=150)

#save
plt.savefig( "1971.png", bbox_inches='tight', dpi=20)  
```

```python
#creates heatmap for each year in the data and saves them in a folder, heat is measured on number of kills per city

#load data and basic manipulation
df = pd.read_csv(r"data_clean.csv", sep="	")
df["year"] = df['start_date'].astype(str).str[:4]
df = df.dropna(subset=['latitude', 'longitude'])

for i in range(2003, 2018, 1):
    #filter for year
    df_year = df.loc[df['year'] == str(i)]
    df_year = df_year[["city","latitude","longitude","nkill"]]
    df_year = df_year.dropna(subset=['latitude', 'longitude',"city"])
    df_year = df_year[df_year.city != 'Unknown']
    
    df_add = pd.DataFrame() 
    df_add["city"]=""
    df_add["latitude"]=""
    df_add["longitude"]=""
    
    #each attack in one city has the same coordinates so we have to add up total number of kills, to determine heat
    #df_add is dataframe with the columns city, latitude, longitude, score
    cities = df_year['city'].unique()
    
    for c in cities:
        a = pd.DataFrame({"city": [df_year.loc[df_year['city'] == c]["city"].iloc[0]],
                 "latitude": [df_year.loc[df_year['city'] == c]["latitude"].iloc[0]],
                 "longitude": [df_year.loc[df_year['city'] == c]["longitude"].iloc[0]]})
        df_add = df_add.append(a)
        
    df_add["nkill"]=""
    df_add = df_add.set_index('city')    
    cities = df_add.index.values
    
    for c in cities:
        df_add.at[c, 'nkill'] = df_year.loc[df_year['city'] == c]["nkill"].sum()
    df_add = df_add.reset_index()
    
    df_add["score"]="" # determines score for each city for size and colour in the map
    df_add.loc[df_add["nkill"] == 0,"score"] = 1000/2
    df_add.loc[(df_add['nkill'] > 0) & (df_add['nkill'] <= 3),"score"] = 1500/2
    df_add.loc[(df_add['nkill'] > 3) & (df_add['nkill'] <= 5),"score"] = 1800/2
    df_add.loc[(df_add['nkill'] > 5) & (df_add['nkill'] <= 10),"score"] = 3000/2
    df_add.loc[(df_add['nkill'] > 10) & (df_add['nkill'] <= 20),"score"] = 4500/2
    df_add.loc[(df_add['nkill'] > 20) & (df_add['nkill'] <= 50),"score"] = 6000/2
    df_add.loc[(df_add['nkill'] > 50) & (df_add['nkill'] <= 80),"score"] = 7000/2
    df_add.loc[(df_add['nkill'] > 80) & (df_add['nkill'] <= 100),"score"] = 8000/2
    df_add.loc[(df_add['nkill'] > 100) & (df_add['nkill'] <= 150),"score"] = 9000/2
    df_add.loc[(df_add['nkill'] > 150) & (df_add['nkill'] <= 250),"score"] = 10000/2
    df_add.loc[(df_add['nkill'] > 250) & (df_add['nkill'] <= int(df_add.nkill.max())),"score"] = 12000/2
    
    
    #necessary parameters for map
    lat = df_add['latitude'].values
    lon = df_add['longitude'].values
    size = df_add['score'].values
    
    #map
    fig = plt.figure(figsize=(75, 75), edgecolor='w')
    m = Basemap(projection='cyl', resolution=None,
                llcrnrlat=-90, urcrnrlat=90,
                llcrnrlon=-180, urcrnrlon=180, )
    m.shadedrelief()
    
    #scatter on map
    m.scatter(lon, lat, 
              latlon = True, 
              c = size, #color, sequence, or sequence of color
              s = size,# #scalar or array_like, shape (n, ), optional
              cmap='Reds',
              alpha=0.5) # The alpha blending value, between 0 (transparent) and 1 (opaque).
    
    #title for map
    plt.title(str(i), fontsize=50)
    
    #save every image in a subfolder
    image_name = "heatmaps_small/"+str(i)+str(".png")
    plt.savefig(image_name, bbox_inches='tight', dpi=20)
    plt.close(fig)
```
