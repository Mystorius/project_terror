```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
df=pd.read_csv('./data_clean.csv',sep='\t')
```

```python
listdf=list(df)
```

```python
listdf
```

```python
df = df.loc[(df['gname'] == 'Taliban') | (df['gname'] == 'Islamic State of Iraq and the Levant (ISIL)') | (df['gname'] == 'Boko Haram')]
```

```python
plt.subplots(figsize=(15,8))
f = sns.countplot(y='region_txt',data=df,order=df['region_txt'].value_counts().index, palette="inferno")
#plt.xticks(rotation=0)
plt.ylabel('')
plt.xlabel('Incident Count')
plt.title('TERRORIST INCIDENTS BY REGION')
plt.show()
```

```python
plt.subplots(figsize=(15,8))
sns.countplot(y='country_txt',data=df,order=df['country_txt'].value_counts().iloc[:10].index, palette="inferno")
#plt.xticks(rotation=0)
plt.ylabel('')
plt.xlabel('Incident Count')
plt.title('TERRORIST INCIDENTS BY COUNTRIES')
plt.show()
```
